# Series-


A Pandas Series is like a column in a table.
It is a one-dimensional array holding data of any type.

## Pandas methods-

series() - Used to create series

combine_first() - Combine two series

count() - Gives no. of none/null elements

size() - Gives no. of elements

Is_unique() - Return true f all elements are unique

```python
# Using list to create Series
import pandas as pd
l=[7,8,9,5,2,6,88,66]
s=pd.Series(l,name="Random numbers")
s
```

```python
# Using dictionary to create Series
dic={"Name":["Anurag kashyap","Ankit Rai","Anupreet mishra"],"Age":[18,17,22]}
s=pd.Series(dic)
s
```

```python
#Using combine_first()
l=[1,2,5,3,8,7,5,8,9]
l2=[9,8,4,2,3,2,6,7,5]
s=pd.Series(l)
s2=pd.Series(l2)
s.combine_first(s2)
```

# Binary operation series-



*   Single Series
*   Multiple Series





### Single Series-

1.   sum()

1.   abs()
2.   prod()


4.   mean()

1.   cov()





```python
# Using sum() operation
l=[2,1,5,2,5,7,8,1,5]
s=pd.Series(l)
s.sum()
```

```python
# Using abs()
l=[2,1,-5,2,-5,7,-8,1,-5]
s=pd.Series(l)
s.abs()
```

```python
# Using prod()
l=[2,1,5,2,5,7,8,1,5]
s=pd.Series(l)
s.prod()
```

```python
# Using mean-> It gives avg
l=[2,1,5,2,5,7,8,1,5]
s=pd.Series(l)
s.mean()
```

## Multiple series-



1.   add()

1.   sub()
2.   mul()


4.   div()



```python
# Using add()
l=[1,2,5,3,8,7,5,8,9]
l2=[9,8,4,2,3,2,6,7,5]
s=pd.Series(l)
s2=pd.Series(l2)
s.add(s2)
```

```python
#using sub
l=[1,2,5,3,8,7,5,8,9]
l2=[9,8,4,2,3,2,6,7,5]
s=pd.Series(l)
s2=pd.Series(l2)
s.sub(s2)
```

```python
#Using Mul
l=[1,2,5,3,8,7,5,8,9]
l2=[9,8,4,2,3,2,6,7,5]
s=pd.Series(l)
s2=pd.Series(l2)
s.mul(s2)
```

```python
#Using div
l=[1,2,5,3,8,7,5,8,9]
l2=[9,8,4,2,3,2,6,7,5]
s=pd.Series(l)
s2=pd.Series(l2)
s.div(s2)
```

## Read CSV(Comma Seprated Value) File-

A simple way to store big data sets is to use CSV files (comma separated files).CSV files contains plain text and is a well know format that can be read by everyone including Pandas.
In our examples we will be using a CSV file called 'data.csv'

### Syntax-  pd.read_csv('File name')

```python
path=("https://raw.githubusercontent.com/codeforamerica/ohana-api/master/data/sample-csv/phones.csv")
s=pd.read_csv(path)
s
```
